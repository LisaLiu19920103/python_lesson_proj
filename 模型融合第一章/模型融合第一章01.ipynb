```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler=MinMaxScaler()
x=min_max_scaler.fit_transform(train_data[feature_cols].values)
x_test=min_max_scaler.transform(test_data[feature_cols].values)
```

```python
from sklearn.model_selection import train_test_split
y=train_data['price'].values
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2)
```

```python
print(train_y)
```

```python
from tensorflow import keras
model=keras.Sequential([
    keras.layers.Dense(300,activation='relu',input_shape=[len(feature_cols)]),
    keras.layers.Dense(200,activation='relu'),
    keras.layers.Dense(100,activation='relu'),
    keras.layers.Dense(1)
])
model.compile(loss='mean_absolute_error',optimizer='Adam')
model.fit(train_x,train_y,batch_size=1024,epochs=200)
```

```python
from sklearn.metrics import mean_absolute_error
print("训练集MAE:",mean_absolute_error(train_y,model.predict(train_x)))
print("验证集MAE:",mean_absolute_error(test_y,model.predict(test_x)))
```

```python
predict_y=model.predict(x_test)
result=pd.DataFrame()
result['SaleID']=test_data['SaleID']
result['price']=predict_y
result['price'][result['price']<11]=11
result.to_csv('G://python_lesson//二手车//price_predict_new.csv',index=False)
```
