### Thinking1	机器学习中的监督学习、非监督学习、强化学习有何区别
答：强化学习是实践-反馈的机制进行学习的，如果实践结果失败给模型负反馈，如果实践结果成功，给模型正反馈，结果的反馈是延后的，不是预先得知的。而监督学习的数据集中带有label标签，模型在训练过程中会通过实际标签与预测标签的loss来不断优化模型中的各项参数。非监督的学习的数据中没有label标签，只有数据特征，而强化学习有奖励值，非监督学习和监督学习中的数据之间是相互独立的，强化学习的数据会有前后的依赖关系。

### Thinking2	什么是策略网络，价值网络，有何区别
答：策略网络中会有一个状态-行动表，这个状态-行动表包括了该游戏中所有的策略，而策略网络就是通过学习该状态-行动表去实现对给定的输入（行为）输出确定的网络。在价值网络中，所有的状态都会有一个相应的分数，而价值网络的目标就是让所有时间序列上的状态对应的价值累计分数最大化。

### Thinking3	请简述MCTS（蒙特卡洛树搜索）的原理，4个步骤Select, Expansion，Simluation，Backpropagation是如何操作的
答：Select：从根节点开始，每一个节点如果存在子节点，则从子节点中选择一个最有价值的子节点，一步步往下搜索，直到走到叶子节点。
    Expansion：扩展，就是对叶子节点来说，去扩展增加它的子节点。
    Simluation：模拟，用快速走子策略去完成一局棋，然后得到胜负结果。
    Backpropagation：根据最终棋局的结果，去反向回传，更新这条链路上的所有结点价值结果。

### Thinking4：	假设你是抖音的技术负责人，强化学习在信息流推荐中会有怎样的作用，如果要进行使用强化学习，都有哪些要素需要考虑
答：在抖音视频推荐中，将用户的所有点击分为分为一个个相等数量的点击链路，将用户在任意一个分块中的视频点击路径看作是马尔科夫链，将最后一个视频用户完全观看看作是环境的正向奖励反馈，将最后一个视频用户未完全观看看作是环境的负向反馈，进行模型训练，在实际训练过程中，很有可能所有的分块中视频完全不一样，导致学习数据矩阵稀疏，也可以将视频用一组特征表示，这样将所有的视频转换为不同的type，去学习用户对不同视频type的喜恶，从而使得在后续推荐过程中增加用户的点击观看概率。

### Thinking5：	在自动驾驶中，如何使用强化学习进行训练，请说明简要的思路
答：自动驾驶在初期的训练数据获取上可能需要一些模拟的游戏还获取，包括模拟设置路标，路障，红绿灯，行人，车辆等等道路信息来训练模型，让车辆能够在根据目前的环境做出正确的反应，模型能够安全到达目的地则获得正向反馈，否则为负向反馈。模型训练后期可能就会需要将真实的带有传感器的车辆，在搭建的训练实景中进行训练，道路数据不再是认为给定的测试数据，而是训练车辆的传感器接收到的实时环境数据，以此来训练自动驾驶模型。

### Action	"Action（五子棋）：棋盘大小 10 * 10采用强化学习（策略价值网络），用AI训练五子棋AI请说明都有哪些模块，不同模块的原理"
答：将五子棋转化为强化学习，其中最重要的三部分分别为：自我对弈、神经网络和蒙特卡洛搜索树（MTCS）。其中在训练过程中通过AI主程序的自我对弈，将当前棋盘的状态与历史记录作为输入，输入深度神经网络进行学习，输出落子概率和价值，即强化学习中的策略网络和价值网络。为了防止神经网络停止优化，引入MTCS，通过MCTS来进行策略优化，在每个状态s处，都会执行蒙特卡洛搜索树给出的策略，直到最终棋局结束，根据棋局的输赢会得到奖励 z 值，该奖励值就是当前的策略评估，然后神经网络根据MTCS学习结果来反向调优自己的参数，从而不断优化参数。
