1.MVC框架指的是什么
答：MVC全名是Model View Controller,Model指的是model层，主要功能是与数据源进行交互，例如对数据库的读取和存储，Controller只得是Controller层，主要对数据的逻辑处理，算法模型的实现主要是在这一层，View指的是View层，包括对数据可视化的实现，是结果的展示层。
2.基于Python的可视化技术都有哪些，你使用过哪些

```python
def create_cloud(f):
    cut_word=word_tokenize(f)
    cut_word=" ".join(cut_word)
    wc_new=wordcloud.WordCloud(max_words=100,height=1200,width=2000)
    wc_new.generate(cut_word)
    wc_new.to_file("wc_new.jpg")
    pl.imshow(wc_new)
    pl.axis("off")
    pl.show


a=pd.read_csv('G:/python_lesson/8127_11403_bundle_archive/Market_Basket_Optimisation.csv',header=None)
#print(a)
print(a.shape)
transaction=[];
words_count={}
for i in range(0,a.shape[0]):
    temp=[]
    #print(temp_list.count)
    for j in range(0,20):
        item=str(a.values[i,j])
        if item!='nan':
            temp.append(item)
            if item not in words_count:
                words_count[item]=1
            else:
                 words_count[item]+=1
                    
    transaction.append(temp)
print(type(transaction))
all_word=" ".join('%s' %item for item in transaction)
create_cloud(all_word)
print(sorted(words_count.items(),key=lambda x:x[1],reverse=True))
#print(transaction.shape)
    
    
    #print(temp_list)
          
```
